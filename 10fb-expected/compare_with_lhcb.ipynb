```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Read the data from Nick
df    = pd.read_csv("btophi_ul.csv")
df1   = df.query("ctau==1")
df10  = df.query("ctau==10")
df100 = df.query("ctau==100")
df100
```

```python
# Rescaling factor (b to B0 or B+, factor 10 more lumi)
f  = 0.344 * math.sqrt(10.)
```

```python
ax=plt.subplot(111)

ax.fill_between(df1['mass'], f*df1['up'], f*df1['down'], facecolor='blue', linewidth=0, alpha=0.3)
ax.plot(df1['mass'], f*df1['central'], color='blue', label='ctau = 1 mm')

ax.fill_between(df10['mass'], f*df10['up'], f*df10['down'], facecolor='green', linewidth=0, alpha=0.3)
ax.plot(df10['mass'], f*df10['central'], color='green', label="ctau = 10 mm")

ax.fill_between(df100['mass'], f*df100['up'], f*df100['down'], facecolor='red', linewidth=0, alpha=0.3)
ax.plot(df100['mass'], f*df100['central'], color='red', label="ctau = 100 mm")

ax.set_xlim([0.5,2])
ax.set_yscale('log')
ax.set_xlabel("Phi mass (GeV)")
ax.set_ylabel("UL (pb) \n BR(B0->Phi X) or BR(B+->Phi X)")
ax.legend()
ax.grid()
```

```python
# Now read the B0 -> K* phi limits from LHCB
neu1   = pd.read_csv("LHCB/LHCb-PAPER-2015-036-supplementary/ctau_1mm.csv",sep=' ')
neu10  = pd.read_csv("LHCB/LHCb-PAPER-2015-036-supplementary/ctau_10mm.csv",sep=' ')
neu100 = pd.read_csv("LHCB/LHCb-PAPER-2015-036-supplementary/ctau_100mm.csv",sep=' ')
neu1
```

```python
ax2=plt.subplot(111)

ax2.fill_between(df1['mass'], f*df1['up'], f*df1['down'], facecolor='blue', linewidth=0, alpha=0.3)
ax2.plot(df1['mass'], f*df1['central'], color='blue', label='ctau = 1 mm')

ax2.fill_between(df10['mass'], f*df10['up'], f*df10['down'], facecolor='green', linewidth=0, alpha=0.3)
ax2.plot(df10['mass'], f*df10['central'], color='green', label="ctau = 10 mm")

ax2.fill_between(df100['mass'], f*df100['up'], f*df100['down'], facecolor='red', linewidth=0, alpha=0.3)
ax2.plot(df100['mass'], f*df100['central'], color='red', label="ctau = 100 mm")

ax2.plot(0.001*neu1['mass'], neu1['ul'], color='blue', label='LHCB ctau = 1 mm', linestyle='dashed')
ax2.plot(0.001*neu10['mass'], neu10['ul'], color='green', label='LHCB ctau = 10 mm', linestyle='dashed')
ax2.plot(0.001*neu100['mass'], neu100['ul'], color='red', label='LHCB ctau = 10 mm', linestyle='dashed')

ax2.set_xlim([0.5,2])
ax2.set_yscale('log')
ax2.set_xlabel("Phi mass (GeV)")
ax2.set_ylabel("UL (pb)")
# ax2.legend()
ax2.grid()
ax2.legend(bbox_to_anchor=(1.05, 0.5))
ax2.set_title("CMS B0->Phi X      LHB B0->Phi K*")
```
